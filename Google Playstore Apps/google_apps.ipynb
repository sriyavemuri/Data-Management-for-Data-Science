## Preprocessing (28 pts)

### [3 pts] Often there are outliers which do not match the overall data type. There is one record in this data where the "Reviews" has value "3.0M" which does not match the rest of the data. Remove that record.

### [4 pts] Remove rows where any of the columns has the value "Varies with device".

```python
for index, row in df_googleplaystore.iterrows():
    for column in df_googleplaystore.columns:
        if row[column] == 'Varies with device':
            df_googleplaystore.drop(index, inplace=True)
            break
df_googleplaystore
```

### [5 pts] The values in the Android version column should be floats. Strip the trailing non-numeric characters from all values (ie. the words " and up"), so the result is a number. If there are multiple decimal places (eg. "x.y.z"), keep only the first two parts (eg "x.y"). For example, the value "4.1 and up" should be changed to "4.1". The value "4.5.6 and up" should be changed to "4.5". The value "5.6.7" should be changed to "5.6".
If there is a range (eg. 5.0 - 8.0), only consider the first number. For example, the value "5.0 - 8.0" should be changed to "5.0". The value "4.0.3 - 7.1.1" should be changed to "4.0".

```python
import re
```

```python
def versioncleaner(version):
    if isinstance(version, (float, int)):
        return version
    match = re.search(r'\d+(\.\d+)?', str(version))
    if match:
        return float(match.group())
    else:
        return np.nan

for index, row in df_googleplaystore.iterrows():
    androidunclean = row['Android Ver']
    androidcleaned = versioncleaner(androidunclean)
    df_googleplaystore.at[index, 'Android Ver'] = androidcleaned

df_googleplaystore['Android Ver'] = pd.to_numeric(df_googleplaystore['Android Ver'], errors='raise')
```

```python
df_googleplaystore
```

### [5 pts] The "Installs" column must have integer values. For values that have commas, remove the commas. For values that have a '+' at the end, remove the '+'. Keep only those rows that have an integer value after these edits.

```python
for index, row in df_googleplaystore.iterrows():
    installs_value = row['Installs']
    cleaned_installs = re.sub(r'[,+]', '', str(installs_value))
    if cleaned_installs.isdigit():
        df_googleplaystore.at[index, 'Installs'] = int(cleaned_installs)
    else:
        df_googleplaystore.at[index, 'Installs'] = pd.NA

df_googleplaystore = df_googleplaystore[df_googleplaystore['Installs'].notna()]
```

```python
df_googleplaystore
```

### [5 pts] For missing rating values, if the number of reviews is less than 100 and installations is less than 50000, remove the row. Else, fill the missing value with the average value (rounded to 2 decimal places) for the Category of that row.

```python
df_googleplaystore = df_googleplaystore[~((df_googleplaystore['Rating'].isna()) & (df_googleplaystore['Reviews'] < 100) & (df_googleplaystore['Installs'] < 50000))]
averageratings = df_googleplaystore.groupby('Category')['Rating'].mean()
for index, row in df_googleplaystore.iterrows():
    if pd.isna(row['Rating']):
        category = row['Category']
        averagerating = round(averageratings[category], 2)
        df_googleplaystore.at[index, 'Rating'] = averagerating
```

```python
df_googleplaystore
```

### [6 pts] Preprocess the Size column to convert the "M" (millions) and "K" (thousands) values into integers. For instance, 8.7M should be converted to 8700000 and 2.4K should be converted to 2400.

```python
def preprocess_size(size):
    if pd.isna(size):
        return size
    elif isinstance(size, int):
        return size
    else:
        match = re.match(r'([\d.]+)([MK]?)', str(size))
        if match:
            numeric_value = float(match.group(1))
            unit = match.group(2)
            if unit == 'M':
                return int(numeric_value * 1000000)
            elif unit == 'K':
                return int(numeric_value * 1000)
        else:
            return None
for index, row in df_googleplaystore.iterrows():
    size = row['Size']
    df_googleplaystore.at[index, 'Size'] = preprocess_size(size)
```

### [11 pts] Extract all "Free" apps from the master data frame. Then write a function that, given a numeric column e.g 'Rating'), will create and return a dataframe for the top 3 free applications in each category based on that column. Call the function on each of these columns:

```python
df_freeapps = df_googleplaystore.loc[df_googleplaystore['Type'] == 'Free']
df_freeapps
```

```python
def top3apps(df_freeapps, column):
    columns = ['Category', 'App', column]
    topthree = []
    grouped = df_freeapps.groupby('Category')
    for category, group in grouped:
        sortedgroup = group.sort_values(by=column, ascending=False)
        top3 = sortedgroup.head(3)[columns]
        topthree.append(top3)
    top3df = pd.concat(topthree)
    return top3df
```

#### a. Rating (gives top 3 most highly rated applications in each category)

```python
df_googleplaystore["Rating"]=(df_googleplaystore["Rating"]).apply(float)
ratingtop3 = top3apps(df_freeapps, 'Rating')
ratingtop3
```

#### b. Installs (gives top 3 most installed applications in each category)

```python
df_googleplaystore["Installs"]=(df_googleplaystore["Installs"]).apply(int)
ratingtop3 = top3apps(df_freeapps, 'Installs')
ratingtop3
```

### [9 pts] In the genre column, break the string of genres into a list. For example, ‘Art & Design; Creativity’ should be [‘Art & Design’, Creativity’].
Count the number of applications per genre and display it using a pie chart.
Hint: Read about DataFrame.explode()

```python
df_googleplaystore
```

```python
genreslist = df_googleplaystore['Genres'].to_list()
genrelist = []
for genre in genreslist:
    genre = str(genre)
    genres = genre.split(';')
    for i in genres:
        genrelist.append(i.strip())
genrelistseries = pd.Series(genrelist)
genrecounts = genrelistseries.value_counts()
genrecounts.head()
```

```python
plt.figure(figsize=(18, 22)) # large figure to see the entire pie
plt.pie(genrecounts, labels=genrecounts.index, autopct='%1.1f%%', startangle=140, rotatelabels=True)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Number of Applications per Genre', size=16)
plt.show()
```
